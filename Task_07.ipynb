```python
# Statistical Business Analysis on Sales Data
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv("sales_dataset.csv")   # TODO: Upload 'sales_dataset.csv' or specify the correct path.

# 1. Summary Statistics
print("\n=== SUMMARY STATISTICS ===")
print(df.describe())

# 2. Correlation Matrix
print("\n=== CORRELATION MATRIX ===")
correlation = df.corr()
print(correlation)

# 3. Hypothesis Test (Marketing Spend vs Revenue)
print("\n=== HYPOTHESIS TEST ===")
pearson_r, p_value = stats.pearsonr(df["Marketing_Spend"], df["Revenue"])
print(f"Pearson Correlation Coefficient: {pearson_r}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("ðŸš€ Reject H0 â†’ Marketing Spend significantly correlates with Revenue")
else:
    print("Fail to reject H0 â†’ No strong evidence of correlation")

# 4. Scatter Plot + Regression Line
plt.scatter(df["Marketing_Spend"], df["Revenue"])
plt.title("Marketing Spend vs Revenue")
plt.xlabel("Marketing Spend")
plt.ylabel("Revenue")

# regression line
slope, intercept, _, _, _ = stats.linregress(df["Marketing_Spend"], df["Revenue"])
plt.plot(df["Marketing_Spend"], slope*df["Marketing_Spend"]+intercept)

plt.show()

# 5. Pairplot of all marketing channels vs revenue (optional)
pd.plotting.scatter_matrix(df, figsize=(10,8))
plt.suptitle("Scatterplot Matrix")
plt.show()
```
